```python
# 'lot_number': (lot_start, lot_end, upland_elevation)

lots = {
    's2r01': (1, 38, 6.5),
    's2r02': (39, 44, 12),
    's2r03': (45, 48, 14),
    's2r04': (49, 64, 14),
    's2r05': (65, 69, 15),
    's2r06': (70, 91, 12),
    's2r07': (92, 117, 11),
    's2r08': (118, 131, 11),
    's2r09': (132, 160, 12),
    's2r10': (161, 173, 10),
    's2r11': (174, 194, 16),
    's2r12': (195, 204, 10),
    's2r13': (205, 227, 10)
}
```

```python
# how this works
# loop through the dictionary and assign the tuple values to: lower, upper, and upland variables
    # query the data to only select the lots within the the range, for this case each section
    # subtract/add the upland elevation as needed
    # append the data frame selection to a list
# once completed through the dictionary, concat all the dataframes to 1 dataframe
# save the whole updated dataframe to a CSV file

ffe_columns = ['FFEMin', 'FFEML', 'FFEMax']
query_columns = ['LotExtID', 'FFEMin', 'FFEML', 'FFEMax']

def create_data_frame(change='-'):
    frames = []
    for k, v in lots.items():
        lower = v[0]
        upper = v[1]
        upland = v[2]
    
        df = inventory_df.query(f'LotExtID >= {lower} and LotExtID <= {upper}')

        if change == '-':
            df.loc[:, ffe_columns] = df[ffe_columns] - upland
        
        elif change == '+':
            df.loc[:, ffe_columns] = df[ffe_columns] + upland
        
        frames.append(df)
    
    inventory_without_upland = pd.concat(frames)
    
    inventory_without_upland.to_csv('de_s2_without_upland.csv', index=False)

    return inventory_without_upland

inventory_without_upland_df = create_data_frame()    
        



# # df[ffe_columns] - 6.5

# frames = [df, inventory_df]
# result = pd.concat(frames)

# result
```
